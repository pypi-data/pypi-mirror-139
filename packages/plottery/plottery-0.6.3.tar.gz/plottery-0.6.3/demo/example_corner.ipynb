Load modules and generate fake data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import plottery
from plottery.statsplots import corner
plottery.__version__
```

For nicer layout:

```python
from plottery.plotutils import update_rcParams
update_rcParams()
```

```python
np.random.seed(1938)

truth = [(11.6*(1+0.1*i), 25*(1+0.25*i)) for i in range(3)]
cov = np.array([[0.4,0.5],[0.5,5]])**2
chains = np.transpose(
    [np.random.multivariate_normal(xo, cov, 100000) for xo in truth],
    axes=(0,2,1))
chains.shape
```

Generate corner plot with several optionals. Feel free to play with them!

```python
figure = corner(
    chains,
    colors=('C3','C1','C0'),
    labels=[r'$p_1$', r'$p_2$'],
    bins=[(20,25),(30,40),(30,40)],
    clevels=(0.68,0.95),
    ls2d=['-','--',':'],
    ls1d=['-','--',':'],
    medians1d=False,
    percentiles1d=False,
    style1d='curve',
    background='filled',
    bcolor=('yellow','orange'),
    alpha=0.25,
    names=['This Work','Previous Work','Earlier Work'],
    limits=[(10,16),(0,60)],
    truths=truth[1],
    truths_in_1d=True,
    truth_color='black',
    smooth=[1.,1.],
    top_labels=False,
    names_kwargs={'bbox_to_anchor': (0.55,0.85)}
)
```
